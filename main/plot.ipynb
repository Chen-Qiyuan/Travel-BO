```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import os
from cycler import cycler
plt.rcParams['axes.prop_cycle'] = cycler('color', plt.colormaps['tab20'].colors)
```

```python
workspace = "Hartmann6"
for workspace in ["Ackley5","Levy4","Hartmann6","Dropwave","Griewank"]:
    # Find all files ending with '_reg.xlsx'
    files = glob.glob(os.path.join(workspace, '*_reg.xlsx'))
    plt.figure(figsize=(10, 6))

    for file in files:
        # Read data (assumes no header)
        df = pd.read_excel(file)[:200]
        
        # Compute cumulative regret
        cumulative_regret = df.cumsum()
        
        # Divide by t (iteration number) to get R/t
        t = cumulative_regret.index + 1  # Index starts at 0  t = 1, 2, ..., T
        average_regret = cumulative_regret.div(t, axis=0)  # Divide row-wise by t
        
        # Compute mean and std across runs
        mean = average_regret.mean(axis=1)
        std = average_regret.std(axis=1) / np.sqrt(10)
        
        # Plotting
        algorithm_name = os.path.basename(file).replace('_reg.xlsx', '')
        MAPPED_NAMES = {
            "TUCB":   "TUCB (1.1)",
            "TUCB15": "TUCB (1.5)",
            "TUCB2":  "TUCB (2.0)",
        }
        plt.plot(mean, label=algorithm_name)
        plt.fill_between(range(len(mean)), mean - std, mean + std, alpha=0.2)

    # Format plot
    plt.xlabel('Round (t)', fontsize = 16)
    plt.ylabel('Log Average Cumulative Regret log$(R_t/t)$',fontsize = 16)
    plt.legend(fontsize=14, loc='lower left')
    plt.xlim(0,200)
    # plt.ylim(0,1.4)
    plt.grid(True)
    plt.yscale("log")
    for tick in plt.gca().get_yticklabels(which='both'):
        tick.set_rotation(90)
        tick.set_ha('center')  # Horizontal alignment
        tick.set_va('center')  # Vertical alignment
    plt.tight_layout()
    plt.savefig(os.path.join("Figures",workspace+"_regret.pdf"), dpi=600)
    plt.show()

    # Find all files ending with '_reg.xlsx'
    files = glob.glob(os.path.join(workspace, '*_travel.xlsx'))

    plt.figure(figsize=(10, 6))

    for file in files:
        # Read data (assumes no header)
        df = pd.read_excel(file)[:200]
        
        # Compute cumulative regret
        cumulative_regret = df.cumsum()
        
        # Divide by t (iteration number) to get R/t
        t = cumulative_regret.index + 1  # Index starts at 0  t = 1, 2, ..., T
        average_regret = cumulative_regret.div(t, axis=0)  # Divide row-wise by t

        # Compute mean and std across runs
        mean = average_regret.mean(axis=1)
        std = average_regret.std(axis=1) / np.sqrt(10)
        
        # Plotting
        algorithm_name = os.path.basename(file).replace('_travel.xlsx', '')
        plt.plot(mean, label=algorithm_name)
        plt.fill_between(range(len(mean)), mean - std, mean + std, alpha=0.2)

    # Format plot
    plt.xlabel('Round (t)',fontsize = 16)
    plt.ylabel('Average Movement Cost $C_t/t$',fontsize = 16)
    plt.legend(fontsize=14, loc='lower left')
    # plt.yscale("log")

    # plt.ylim(-0.5,25)
    plt.xlim(0,200)

    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join("Figures",workspace+"_cost.pdf"), dpi=600)
    plt.show()
```
